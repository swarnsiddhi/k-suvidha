```python
import requests

def get_weather_forecast(city, api_key):
    base_url = "https://api.weatherbit.io/v2.0/forecast/daily"
    params = {
        "city": city,
        "key": api_key,
        "days": 7
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        for day in data["data"]:
            print(f"Date: {day['datetime']}, Temp: {day['temp']}°C, Humidity: {day['rh']} Description: {day['weather']['description']}")
    else:
        print("Error:", response.status_code, response.text)

# Replace with your API key and desired city
api_key = "1e441d8391794534b9bbf4e9ca6dd4ea"
city = "Mumbai"
get_weather_forecast(city, api_key)
```
